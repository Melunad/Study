# Задача 3 Простой калькулятор

```python
sttr = input()
w = sttr.split(" ")

if w[1] == "+":
  print(int(w[0]) + int(w[2]))
elif w[1] == "-":
  print(int(w[0]) - int(w[2]))
elif w[1] == "/":
  if(w[2] != '0'):
     print(int(w[0]) / int(w[2]))
  else:
    print("На 0 не делим")
elif w[1] == "//":
    if(w[2] != "0"):
        print(int(w[0]) // int(w[2]))
    else:
        print("На 0 не делим")
elif w[1] == "pow":
    print(int(w[0]) ** int(w[2]))
elif w[1] == "abs":
    print(abs(int(w[0])), " ", abs(int(w[2])))
else:
  print("error")
```

# Задача 4 Пока не ноль


```python
sum = int(input("Ввод прекратиться когда сумма будет ровняться нулю \n", ))
square = sum ** 2

while(sum != 0):
  buf = int(input())
  sum += buf
  square += buf ** 2

print("Ответ: ", square)
```

# Задача 5 Последованость длиною N


```python
N = int(input("число не отрицательное \n"))
lst = []
count = 1

while len(lst) != N:

  buf = count
  while buf and len(lst) != N:
    lst.append(count)
    buf -= 1
  count += 1


print(lst)

```

```python
class Fail:

  name = " "
  permission = []
  xcod = {
      "write"   : "w",
      "read"    : "r",
      "execute" : "x",
          }


  def __init__(self, name, permission):
    self.name = name
    self.permission = permission

  def check(self, permission):
    f = 1
    for i in range(len(permission)):
      if self.xcod[permission[i]] in self.permission:
        f = f * 1
      else:
        f = f * 0
    if f == 1:
      return "OK"
    else:
      return "Access denied"
  def getName(self):
    return self.name

  pass

#-----------------------------------------

N = int(input("Кол файлов "))

fails = []

for i in range(N):
  fails.append(Fail(input("Name = "),(input("\n w r x = ")).split(" ")))

for j in range(int(input("Количество запросов "))):
  name = input("Name = ")
  for i in range(N):
    if name == fails[i].getName():
        print(fails[i].check(input("write, read or execute ").split(" ")))
```
