### Задача 4.2. Файлы. Обработка данных.
Дан файл f, компоненты которого являются действительными числами. Найти сумму наибольшего и наименьшего из значений компонент.

```python
bufmin = math.inf
bufmax = 0
buf = " "
counts = open('counts.txt', "r+")
raz = len(counts.readlines())
counts.seek(0)

for i in range(raz):
  buf = counts.readline().split(" ")
  for j in range(len(buf)):
    bufmax = max(bufmax, int(buf[j]))
    bufmin = min(bufmin, int(buf[j]))

print(bufmin + bufmax)



#print(len(counts.readline()))
```
